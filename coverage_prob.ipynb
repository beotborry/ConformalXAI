```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = "n02100236_390.JPEG"
expl_method = "GradCAM"

for n_sample in [1000]:
    for eval_n_sample in [1000]:
        with open(f"results/20230205_pred_orig_eval_orig_transform_both_sign_all_reduction_sum/{img}_expl_{expl_method}_sample_{n_sample}_eval_sample_{eval_n_sample}_sigma_0.05_seed_0_results.npy", "rb") as f:
            results = np.load(f, allow_pickle=True)
        print(np.mean(results[0]['coverage_prob']))
        plt.hist(results[0]['coverage_prob'].flatten(), label = f"cal: {n_sample}, eval: {eval_n_sample}", alpha=0.5)

plt.legend()
```
